```python
import numpy as np 
import pickle
import matplotlib.pyplot as plt
import matplotlib.colors as mcol
```

```python
data1 = pickle.load(open('pickles/full_trun_pen_nopen/full_nopen.p', 'rb'))
data2 = pickle.load(open('pickles/full_trun_pen_nopen/full_pen.p', 'rb'))
data3 = pickle.load(open('pickles/full_trun_pen_nopen/trunc_nopen.p', 'rb'))
data4 = pickle.load(open('pickles/full_trun_pen_nopen/trunc_pen.p', 'rb'))
```

```python
test = pickle.load(open('pickles/initial_training.p', 'rb'))
```

```python
x = 0
titles = ['Total Reward', 'Policy Loss', 'Value Loss', 'Trial Length']
fig, ax = plt.subplots(2,1,sharex=True)
ax[0].plot(data4[x], label = 'tr_p')
ax[0].plot(data3[x], label = 'tr_n')
ax[0].set_title(titles[x])
ax[0].legend(ncol=1, bbox_to_anchor = (1.2, .5))

ax[1].plot(data2[x], label='f_p')
ax[1].plot(data1[x], label='f_n')

ax[1].legend(ncol=1, bbox_to_anchor = (1.2, 0.5))
plt.show()
```

```python
x = 0
titles = ['Total Reward', 'Policy Loss', 'Value Loss', 'Trial Length']
plt.figure()
plt.plot(test[x], label=titles[x])
plt.legend(bbox_to_anchor=(1.2,0.5))
plt.show()
```
